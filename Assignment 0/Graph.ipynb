```python
import matplotlib.pyplot as plt

fig,ax=plt.subplots()

ax.axhline(0,color='black',linewidth=5)
ax.axvline(0,color='black',linewidth=5)
plt.grid(True)

ax.set_xlim(0,10)
ax.set_ylim(0,10)

ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_title("X and Y Axes")

a=[5,5,5,6,4,9,2,6,5,5,5]
b=[7,4,1,10,8,2,5,8,9,6,3]
plt.plot(a)
plt.plot(b)

plt.show()
```
