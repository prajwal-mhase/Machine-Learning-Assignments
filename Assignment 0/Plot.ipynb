```python
import matplotlib.pyplot as plt

fig,ax=plt.subplots()

ax.axhline(0,color='black',linewidth=5)
ax.axvline(0,color='black',linewidth=5)
plt.grid(True)

ax.set_xlim(0,10)
ax.set_ylim(0,10)

ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_title("X and Y Axes")

a=[3,6,8,3,9,1,7,3,4,5]
plt.plot(a)

plt.show()
```
