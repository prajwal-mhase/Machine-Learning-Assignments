```python
data=data.drop(['S.No.'],axis=1)
data.info()
```

```python
data['Brand']=data.Name.str.split().str.get(0)
data['Model']=data.Name.str.split().str.get(1) +" "+ data.Name.str.split().str.get(2)
data[['Name','Brand','Model']]
```

```python
data = {
    'Fuel_Type': ['Petrol', 'Diesel', 'Petrol', 'Diesel', 'CNG', 'Petrol', 'CNG', 'Diesel'],
    'Price': [500000, 700000, 450000, 720000, 300000, 480000, 310000, 690000]
}
df = pd.DataFrame(data)
avg_price = df.groupby('Fuel_Type')['Price'].mean().sort_values()
plt.figure(figsize=(8,5))
avg_price.plot(kind='bar', color='skyblue')
plt.title('Average Price by Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Average Price')
plt.xticks(rotation=0)
plt.show()
```

```python
data = {
    'Fuel_Type': ['Petrol', 'Diesel', 'Petrol', 'Diesel', 'CNG', 'Petrol', 'CNG', 'Diesel']
}
df = pd.DataFrame(data)
fuel_counts = df['Fuel_Type'].value_counts()
plt.figure(figsize=(6,6))
fuel_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['gold', 'lightgreen', 'lightblue'])
plt.title('Distribution of Fuel Types')
plt.ylabel('')
plt.show()
```

```python
data = {
    'Price': [500000, 700000, 450000, 720000, 300000, 480000, 310000, 690000]
}
df = pd.DataFrame(data)

plt.figure(figsize=(8,5))
plt.hist(df['Price'], bins=5, color='skyblue', edgecolor='black')
plt.title('Histogram of Price')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```
